## Water bucket problem using Depth first Search and Breadth first Search

```python
import collections

def Start():
    #Declaration Section
    b1 = int(input("Size of Bucket A: "))
    b2 = int(input("Size of Bucket B: "))
    bucketsSize = [b1,b2]
    
    target = int(input("Target measure: "))
    
    response = input("Type of Search: Depth of Search(D) Breath of Search(B) ")
    if response == "D" or response == "d":
        searchtype = True
    else:
        searchtype = False
    
    pathVisited = collections.deque() # create queue that can enqueue form both sides
    pathVisited.append([[0,0]])
    solutionPaths = []
    #--Declaration Section#
    
    #Algorithm section
    while len(pathVisited)!=0:
        path = pathVisited.popleft()
        currentNode = path[-1]

        solvednode = ConditionMatch(target,currentNode)
        if solvednode!=[]:
            path.append(solvednode)
            solutionPaths.append(path)
            continue
        
        pathList = PossiblePathList(bucketsSize,currentNode,path)
        for i in pathList:
            if searchtype:
                # For Depth first Search, paths returen by PossiblePathList add to left side of the Queue
                # DeQueue is done form left hand side of the queue, newly add paths get first
                # meas going down through the tree structure
                pathVisited.appendleft(i) 
            else:
                # For Breadth first Search, paths returen by PossiblePathList add to right side of the Queue
                # DeQueue is done form left hand side of the queue, early add paths get first
                # meas going level by levl in the tree structure
                pathVisited.append(i) 
    #--Algorithm section
    
    #Disply Output 
    if len(solutionPaths) > 0:
        print("\nTotal Solution path",len(solutionPaths),end="\n\n")
    
        for path in solutionPaths:
            print(path)

        solutionPaths.sort(key=len)
        print("\nShortet Path to target: ",end="")

        for index,node in enumerate(solutionPaths[0],start=0):
            if index != 0:
                print(">>",end="")
            print(node,end="")
    else:
        print("No path to achieve the target")
     #--Disply OutPut 
    
def PossiblePathList(bucketsSize,currentNode,pathVisited):
    paths = []
    
    #empty bucket 1
    nextStep = []
    nextStep.append(0)
    nextStep.append(currentNode[1])
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
    
    #empty bucket 2
    nextStep = []
    nextStep.append(currentNode[0])
    nextStep.append(0)
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)  
        
    #fully-fill bucket 1 
    nextStep = []
    nextStep.append(bucketsSize[0])
    nextStep.append(currentNode[1])
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
        
    #fully-fill bucket 2         
    nextStep = []
    nextStep.append(currentNode[0])
    nextStep.append(bucketsSize[1])
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
        
    #pour bucket 1 to bucket 2 
    nextStep = []
    nextStep.append(max(0,currentNode[0]-(bucketsSize[1]-currentNode[1])))
    nextStep.append(currentNode[1]+(currentNode[0]-nextStep[0]))
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
    
    #pour bucket 2 to bucket 1
    nextStep = []
    nextStep.append(max(0,(currentNode[1]-(bucketsSize[0]-currentNode[0]))))
    nextStep.insert(0,currentNode[0]+(currentNode[1]-nextStep[0]))
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
    
    # new path is created for each possible next-node, by appending next-node to visited path
    result =[]
    for path in paths:
        temp = list(pathVisited)
        temp.append(path)
        result.append(temp)
    return result
    
def IsNextStepValid(nextStep,pathVisited):
    #check is nextstep is in the visited path list if yes return false.
    result = False
    for node in pathVisited:
        if node == nextStep:
            return False
        else:
            result = True
            
    return result
        
    
def ConditionMatch(target,currentNode):
    # check if the target is achievable, if achievable crate end transaction
    if target == currentNode[0]:
        return [target,0]
    elif target == currentNode[1]:
        return [0,target]
    else:
        return []
        
Start()
```
