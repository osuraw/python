```python
import collections

def Start():
    #Declaration Section
    b1 = int(input("Size of Bucket A: "))
    b2 = int(input("Size of Bucket B: "))
    bucketsSize = [b1,b2]
    
    target = int(input("Target measure: "))
    
    response = input("Type of Search: Depth of Search(D) Breath of Search(B) ")
    if response == "D";
        searchtype = False 
    
    pathVisited = collections.deque() # create queue that can enqueue form both sides
    pathVisited.append([[0,0]])
    solutionPaths = []
    #--Declaration Section#
    
    #Algorithm section
    while len(pathVisited)!=0:
        path = pathVisited.popleft()
        currentNode = path[-1]

        solvednode = ConditionMatch(target,currentNode)
        if solvednode!=[]:
            path.append(solvednode)
            solutionPaths.append(path)
            continue
        
        pathList = PossiblePathList(bucketsSize,currentNode,path)
        for i in pathList:
            if searchtype:
                pathVisited.appendleft(i)
            else:
                pathVisited.append(i)
    #--Algorithm section
    
    #Disply Output 
    if len(solutionPaths) > 0:
        print("Total Solution path",len(solutionPaths))
    
        for path in solutionPaths:
            print(path)

        solutionPaths.sort(key=len)
        print("Shortet Path to target: ",end="")

        for index,node in enumerate(solutionPaths[0],start=0):
            if index != 0:
                print(">>",end="")
            print(node,end="")
    else:
        print("No path to achieve the target")
     #--Disply OutPut 
    
def PossiblePathList(bucketsSize,currentNode,pathVisited):
    paths = []
    
    #empty bucket 1
    nextStep = []
    nextStep.append(0)
    nextStep.append(currentNode[1])
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
    
    #empty bucket 2
    nextStep = []
    nextStep.append(currentNode[0])
    nextStep.append(0)
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)  
        
    #fully-fill bucket 1 
    nextStep = []
    nextStep.append(bucketsSize[0])
    nextStep.append(currentNode[1])
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
        
    #fully-fill bucket 2         
    nextStep = []
    nextStep.append(currentNode[0])
    nextStep.append(bucketsSize[1])
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
        
    #pour bucket 1 to bucket 2 
    nextStep = []
    nextStep.append(max(0,currentNode[0]-(bucketsSize[1]-currentNode[1])))
    nextStep.append(currentNode[1]+(currentNode[0]-nextStep[0]))
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
    
    #pour bucket 2 to bucket 1
    nextStep = []
    nextStep.append(max(0,(currentNode[1]-(bucketsSize[0]-currentNode[0]))))
    nextStep.insert(0,currentNode[0]+(currentNode[1]-nextStep[0]))
    if IsNextStepValid(nextStep,pathVisited):
        paths.append(nextStep)
    
    result =[]
    for path in paths:
        temp = list(pathVisited)
        temp.append(path)
        result.append(temp)
    return result
    
def IsNextStepValid(nextStep,pathVisited):
    result = False
    for node in pathVisited:
        if node == nextStep:
            return False
        else:
            result = True
            
    return result
        
    
def ConditionMatch(target,currentNode):
    if target == currentNode[0]:
        return [target,0]
    elif target == currentNode[1]:
        return [0,target]
    else:
        return []
        
Start()
```
